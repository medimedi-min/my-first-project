```python
import pandas as pd
import numpy as np

print("Pandas version:",pd.__version__)
print("Numpy version:, np._version_")
# 당신의 투자 관심을 '데이터 표'로 만듭니다.
my_portfolio = {
    'Ticker': ['SPYM', 'Alphabet', 'IONQ'],
    'Price': [79.26, 256.32, 60.86],
    'Reason': ['S&P500', 'Ai', 'Quantum']
}

# 이 '정보 뭉치'를 Pandas의 '데이터 표(DataFrame)'로 만듭니다.
df = pd.DataFrame(my_portfolio)

# '데이터 표'를 화면에 출력합니다.
print(df)
```
